````python
# TODO: Ù…ØªØ¯ Ø²ÛŒØ± Ø±Ø§ Ú©Ø§Ù…Ù„ Ú©Ù†ÛŒØ¯.


def impute_missing_data(
    df: pd.DataFrame,
    strategy_dict: dict = None,
    auto_impute: bool = True
) -> pd.DataFrame:
    """
    Ù¾Ø± Ú©Ø±Ø¯Ù† Ù…Ù‚Ø§Ø¯ÛŒØ± Ú¯Ù…Ø´Ø¯Ù‡ Ø¨Ø§ Ø¯Ùˆ Ø±ÙˆØ´ Ø®ÙˆØ¯Ú©Ø§Ø± ÛŒØ§ Ø§Ø³ØªØ±Ø§ØªÚ˜ÛŒâ€ŒÙ‡Ø§ÛŒ Ù…Ø´Ø®Øµ
    
    Ù¾Ø§Ø±Ø§Ù…ØªØ±Ù‡Ø§:
        df (pd.DataFrame): Ø¯ÛŒØªØ§ÙØ±ÛŒÙ… ÙˆØ±ÙˆØ¯ÛŒ
        strategy_dict (dict): Ø¯ÛŒÚ©Ø´Ù†Ø±ÛŒ Ø§Ø³ØªØ±Ø§ØªÚ˜ÛŒâ€ŒÙ‡Ø§ÛŒ Ù¾Ø± Ú©Ø±Ø¯Ù† (Ø§Ø®ØªÛŒØ§Ø±ÛŒ)
        auto_impute (bool): Ù¾Ø± Ú©Ø±Ø¯Ù† Ø®ÙˆØ¯Ú©Ø§Ø± (Ù¾ÛŒØ´â€ŒÙØ±Ø¶: True)
    
    Ø¨Ø§Ø²Ú¯Ø´Øª:
        pd.DataFrame: Ø¯ÛŒØªØ§ÙØ±ÛŒÙ… Ø¨Ø¯ÙˆÙ† Ù…Ù‚Ø§Ø¯ÛŒØ± Ú¯Ù…Ø´Ø¯Ù‡
    """
    df_clean = df.copy()


    # Ù¾Ø± Ú©Ø±Ø¯Ù† Ø®ÙˆØ¯Ú©Ø§Ø±
    if auto_impute:
        # Ù¾Ø± Ú©Ø±Ø¯Ù† Ø³ØªÙˆÙ†â€ŒÙ‡Ø§ÛŒ Ø¹Ø¯Ø¯ÛŒ Ø¨Ø§ Ù…ÛŒØ§Ù†Ú¯ÛŒÙ†
        num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
        for col in num_cols:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
        
        # Ù¾Ø± Ú©Ø±Ø¯Ù† Ø³ØªÙˆÙ†â€ŒÙ‡Ø§ÛŒ ØºÛŒØ±Ø¹Ø¯Ø¯ÛŒ Ø¨Ø§ Ù…Ø¯
        cat_cols = df_clean.select_dtypes(include=['object', 'category']).columns
        for col in cat_cols:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    
    # Ù¾Ø± Ú©Ø±Ø¯Ù† Ø¨Ø§ Ø§Ø³ØªØ±Ø§ØªÚ˜ÛŒâ€ŒÙ‡Ø§ÛŒ Ù…Ø´Ø®Øµ
    elif not auto_impute and strategy_dict is not None:
        for col, strategy in strategy_dict.items():
            if col not in df_clean.columns:
                # raise ValueError(f"Column '{col}' not found in the DataFrame.")
                print( f"Column '{col}' not found in the DataFrame.")
                continue  # Ø±Ø¯ Ú©Ø±Ø¯Ù† Ø³ØªÙˆÙ†â€ŒÙ‡Ø§ÛŒ Ù†Ø§Ù…ÙˆØ¬ÙˆØ¯
            
            # Ø§Ø³ØªØ±Ø§ØªÚ˜ÛŒâ€ŒÙ‡Ø§ÛŒ Ø¹Ø¯Ø¯ÛŒ
            if strategy == 'mean':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
                continue
            elif strategy == 'median':
                df_clean[col] = df_clean[col].fillna(df_clean[col].median())
                continue
            # Ø§Ø³ØªØ±Ø§ØªÚ˜ÛŒâ€ŒÙ‡Ø§ÛŒ ØºÛŒØ±Ø¹Ø¯Ø¯ÛŒ
            elif strategy == 'mode':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
                continue
            elif strategy == 'unknown':
                df_clean[col] = df_clean[col].fillna('unknown')
                continue
            elif strategy == "drop_na":
                df_clean = df_clean.dropna(subset=[col])
                continue
            # Ø§Ø³ØªØ±Ø§ØªÚ˜ÛŒ KNN (ÙÙ‚Ø· Ø¨Ø±Ø§ÛŒ Ø¯Ø§Ø¯Ù‡â€ŒÙ‡Ø§ÛŒ Ø¹Ø¯Ø¯ÛŒ)
            elif strategy == 'knn':
                if pd.api.types.is_numeric_dtype(df[col]) and df[col].nunique() >= 10:
                    scaler = StandardScaler()
                    scaled_data = scaler.transform(df_clean[col])
                    imputer = KNNImputer(n_neighbors=3)
                    scaled_data = imputer.transform(scaled_data)
                    df_clean.loc[:, col] = scaler.inverse_transform(scaled_data)

    
    return df_clean
````

````python
def data_balancing(methods: dict,models:dict, X_train_encoded, X_val_encoded, X_test_encoded, 
                   y_train_encoded, y_val_encoded, y_test_encoded):

    
    # Ø¯ÛŒÚ©Ø´Ù†Ø±ÛŒ Ø¨Ø±Ø§ÛŒ Ø°Ø®ÛŒØ±Ù‡ Ù†ØªØ§ÛŒØ¬
    results = {}
    
    # Ø¨Ø±Ø±Ø³ÛŒ ØªÙˆØ²ÛŒØ¹ Ø§ÙˆÙ„ÛŒÙ‡ Ø¯Ø§Ø¯Ù‡ Train
    print("ØªÙˆØ²ÛŒØ¹ Ø§ÙˆÙ„ÛŒÙ‡ Y_Train:", (y_train_encoded.shape))
    
    for method_name, method in methods.items():
        print(f"\n--- Ø¯Ø± Ø­Ø§Ù„ Ù¾Ø±Ø¯Ø§Ø²Ø´ Ø±ÙˆØ´ {method_name} ---")
        
        # Ø§Ø¹Ù…Ø§Ù„ Ù†Ù…ÙˆÙ†Ù‡â€ŒÚ¯ÛŒØ±ÛŒ ÙÙ‚Ø· Ø±ÙˆÛŒ Train
        X_train_encoded, y_train_encoded = method.fit_resample(X_train_encoded, y_train_encoded)
        print(f"ØªÙˆØ²ÛŒØ¹ Ø¬Ø¯ÛŒØ¯ Train ({method_name}):", Counter(y_train_encoded))
        
        # Ø°Ø®ÛŒØ±Ù‡ Ù†ØªØ§ÛŒØ¬ Ø¨Ø±Ø§ÛŒ Ø§ÛŒÙ† Ø±ÙˆØ´
        results[method_name] = {}
        
        for model_name, model in models.items():
            print(f"\nØ¢Ù…ÙˆØ²Ø´ Ù…Ø¯Ù„ {model_name}...")
            
            # Ø¢Ù…ÙˆØ²Ø´ Ù…Ø¯Ù„
            model.fit(X_train_encoded, y_train_encoded)
            
            # Ù¾ÛŒØ´â€ŒØ¨ÛŒÙ†ÛŒ Ùˆ Ø§Ø­ØªÙ…Ø§Ù„â€ŒÙ‡Ø§
            y_train_pred = model.predict(X_train_encoded)
            y_train_proba = model.predict_proba(X_train_encoded)[:, 1]  # Ø¨Ø±Ø§ÛŒ Ú©Ù„Ø§Ø³ Ù…Ø«Ø¨Øª
            y_val_pred = model.predict(X_val_encoded)
            y_val_proba = model.predict_proba(X_val_encoded)[:, 1]  # Ø¨Ø±Ø§ÛŒ Ú©Ù„Ø§Ø³ Ù…Ø«Ø¨Øª
            
            y_test_pred = model.predict(X_test_encoded)
            y_test_proba = model.predict_proba(X_test_encoded)[:, 1]
            
            # Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø¹ÛŒØ§Ø±Ù‡Ø§ÛŒ Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ Ø¨Ø±Ø§ÛŒ Train
            print(f"""
                 -------------Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø¹ÛŒØ§Ø±Ù‡Ø§ÛŒ Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ Ø¨Ø±Ø§ÛŒ Train : {model_name} + {method_name}---------------------- 
                'AUC-ROC': {roc_auc_score(y_train_encoded, y_train_proba)},
                <---classification_report--->
                {classification_report(y_train_encoded, y_train_pred)}
                

     
                 -----------Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø¹ÛŒØ§Ø±Ù‡Ø§ÛŒ Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ Ø¨Ø±Ø§ÛŒ Validation : {model_name} + {method_name}-------------
                'AUC-ROC': {roc_auc_score(y_val_encoded, y_val_proba)},
                <---classification_report--->
                {classification_report(y_val_encoded, y_val_pred)}
                
        
 
                --------------Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø¹ÛŒØ§Ø±Ù‡Ø§ÛŒ Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ Ø¨Ø±Ø§ÛŒ Test : {model_name} + {method_name}--------------------- 
                'AUC-ROC': {roc_auc_score(y_test_encoded, y_test_proba)},

                <---classification_report--->
                {classification_report(y_test_encoded, y_test_pred)}
__________________________________________________________________________________________________________________________
            """)

    return results
````
