````python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches  # برای ساخت مربعی‌های رنگی در راهنما
%matplotlib inline 
````

````python
# TODO: متد زیر را برای رسم توزیع داده ها کامل کنید.
def plot_numeric_distributions(df: pd.DataFrame,numeric_cols: list):
    df = df.copy()

    # تنظیمات رنگ و ظاهر
    fraud_labels = ['Legitimate (0)', 'Fraud (1)']
    colors = sns.color_palette('Set1', n_colors=2)  # ۲ رنگ برای دو کلاس
    
    # تنظیمات نمودار
    plt.figure(figsize=(20, 15))
    plt.suptitle('Distribution of numerical features based on fraud class', fontsize=18, y=1.02)

    # محاسبه تعداد سطرها و ستون‌ها برای شبکه نمودارها
    n_cols = 3
    n_rows = (len(numeric_cols) // n_cols) + 1

    # رسم هیستوگرام گروهی برای هر ویژگی
    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(data=df, x=col, hue='Fraud_Label', bins=20, palette=colors, alpha=0.6, kde=True)
        plt.title(f'Distribution {col}', fontsize=12)
        plt.xlabel(col, fontsize=10)
        plt.ylabel('Number', fontsize=10)

    # افزودن راهنمای سفارشی
    patches = [mpatches.Patch(color=colors[i], label=fraud_labels[i]) for i in range(2)]
    plt.legend(handles=patches, title='Transaction status', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
````

````python
# انتخاب خودکار ستون‌های عددی (به جز Fraud_Label)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_cols = [col for col in numeric_cols if col != 'Fraud_Label']  # حذف خودکار ستون هدف
plot_numeric_distributions(df, numeric_cols)
````

````python
import numpy as np
from scipy import stats


numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numeric_cols = [col for col in numeric_cols if col != "Fraud_Label"]

def detect_outliers(df,numeric_cols):

    for col in numeric_cols:
        # TODO: شناسایی داده های پرت به کمک IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers_iqr = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
        
        # TODO: شناسایی داده های پرت به کمک Z-Score
        z_scores = np.abs(stats.zscore(df[col]))
        outliers_zscore = (z_scores > 3).sum()
        print(f"Outliers using IQR for column {col}: {outliers_iqr}")
        print(f"Outliers using Z-Score for column {col}: {outliers_zscore}")
        print()


# فراخوانی تابع و نمایش نتایج
detect_outliers(df,numeric_cols)
````

        <br><br>
        درک و نوشتن این تابع به صورت گام به گام کمک می‌کند تا با روش‌های مختلف پر کردن داده‌های گمشده آشنا شوید و بتوانید آن‌ها را به صورت بهینه پیاده‌سازی کنید.
        <br><br>
        1. <strong>ورودی‌ها:</strong>
            <li><strong>df:</strong> دیتافریم ورودی که شامل داده‌های اصلی شما است و ممکن است برخی مقادیر گمشده داشته باشد.</li>
            <li><strong>strategy_dict:</strong> دیکشنری که برای هر ستون یک استراتژی پر کردن (مثل میانگین، میانه، مد و غیره) را مشخص می‌کند. این ورودی فقط در صورتی استفاده می‌شود که بخواهید استراتژی‌های خاصی برای هر ستون انتخاب کنید.</li>
            <li><strong>auto_impute:</strong> یک پارامتر بولی که اگر مقدار آن <code>True</code> باشد، داده‌ها به طور خودکار پر می‌شوند. برای ستون‌های عددی با میانگین و برای ستون‌های رشته‌ای با مد پر می‌شود. در غیر این صورت، از <code>strategy_dict</code> استفاده می‌شود.</li>
        <br>
        2. <strong>پر کردن خودکار (اگر <code>auto_impute=True</code>):</strong>
            <li>در این بخش، اگر شما <code>auto_impute</code> را <code>True</code> قرار دهید، تمامی ستون‌های عددی به صورت خودکار با میانگین پر می‌شوند و ستون‌های رشته‌ای با مد پر خواهند شد. این روش سریع و راحت است اما ممکن است برای همه داده‌ها مناسب نباشد.</li>
        <br>
        3. <strong>پر کردن با استراتژی مشخص (اگر <code>strategy_dict</code> داده شده):</strong>
            <li>در صورتی که بخواهید کنترل بیشتری بر روی پر کردن مقادیر گمشده داشته باشید، می‌توانید از <code>strategy_dict</code> استفاده کنید. به این صورت که برای هر ستون، یک استراتژی خاص انتخاب می‌کنید. این استراتژی‌ها می‌توانند شامل مواردی مانند <code>mean</code> (میانگین)، <code>median</code> (میانه)، <code>mode</code> (مد)، یا حتی استفاده از الگوریتم‌هایی مانند <code>knn</code> (کوتاه‌ترین همسایگی) برای پر کردن داده‌ها باشد.</li>

````python
# TODO: متد زیر را کامل کنید.


def impute_missing_data(
    df: pd.DataFrame,
    strategy_dict: dict = None,
    auto_impute: bool = True
) -> pd.DataFrame:
    """
    پر کردن مقادیر گمشده با دو روش خودکار یا استراتژی‌های مشخص
    
    پارامترها:
        df (pd.DataFrame): دیتافریم ورودی
        strategy_dict (dict): دیکشنری استراتژی‌های پر کردن (اختیاری)
        auto_impute (bool): پر کردن خودکار (پیش‌فرض: True)
    
    بازگشت:
        pd.DataFrame: دیتافریم بدون مقادیر گمشده
    """
    df_clean = df.copy()


    # پر کردن خودکار
    if auto_impute:
        # پر کردن ستون‌های عددی با میانگین
        num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
        for col in num_cols:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
        
        # پر کردن ستون‌های غیرعددی با مد
        cat_cols = df_clean.select_dtypes(include=['object', 'category']).columns
        for col in cat_cols:
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
    
    # پر کردن با استراتژی‌های مشخص
    elif not auto_impute and strategy_dict is not None:
        for col, strategy in strategy_dict.items():
            if col not in df_clean.columns:
                # raise ValueError(f"Column '{col}' not found in the DataFrame.")
                print( f"Column '{col}' not found in the DataFrame.")
                continue  # رد کردن ستون‌های ناموجود
            
            # استراتژی‌های عددی
            if strategy == 'mean':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
                continue
            elif strategy == 'median':
                df_clean[col] = df_clean[col].fillna(df_clean[col].median())
                continue
            # استراتژی‌های غیرعددی
            elif strategy == 'mode':
                df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
                continue
            elif strategy == 'unknown':
                df_clean[col] = df_clean[col].fillna('unknown')
                continue
            elif strategy == "drop_na":
                df_clean = df_clean.dropna(subset=[col])
                continue
            # استراتژی KNN (فقط برای داده‌های عددی)
            elif strategy == 'knn':
                if pd.api.types.is_numeric_dtype(df[col]) and df[col].nunique() >= 10:
                    scaler = StandardScaler()
                    scaled_data = scaler.transform(df_clean[col])
                    imputer = KNNImputer(n_neighbors=3)
                    scaled_data = imputer.transform(scaled_data)
                    df_clean.loc[:, col] = scaler.inverse_transform(scaled_data)

    
    return df_clean
````

````python
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.combine import SMOTETomek, SMOTEENN
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier
from collections import Counter
from sklearn.metrics import  roc_auc_score, classification_report
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from lightgbm import LGBMClassifier
````

````python
def data_balancing(methods: dict,models:dict, X_train_encoded, X_val_encoded, X_test_encoded, 
                   y_train_encoded, y_val_encoded, y_test_encoded):

    
    # دیکشنری برای ذخیره نتایج
    results = {}
    
    # بررسی توزیع اولیه داده Train
    print("توزیع اولیه Y_Train:", (y_train_encoded.shape))
    
    for method_name, method in methods.items():
        print(f"\n--- در حال پردازش روش {method_name} ---")
        
        # اعمال نمونه‌گیری فقط روی Train
        X_train_encoded, y_train_encoded = method.fit_resample(X_train_encoded, y_train_encoded)
        print(f"توزیع جدید Train ({method_name}):", Counter(y_train_encoded))
        
        # ذخیره نتایج برای این روش
        results[method_name] = {}
        
        for model_name, model in models.items():
            print(f"\nآموزش مدل {model_name}...")
            
            # آموزش مدل
            model.fit(X_train_encoded, y_train_encoded)
            
            # پیش‌بینی و احتمال‌ها
            y_train_pred = model.predict(X_train_encoded)
            y_train_proba = model.predict_proba(X_train_encoded)[:, 1]  # برای کلاس مثبت
            y_val_pred = model.predict(X_val_encoded)
            y_val_proba = model.predict_proba(X_val_encoded)[:, 1]  # برای کلاس مثبت
            
            y_test_pred = model.predict(X_test_encoded)
            y_test_proba = model.predict_proba(X_test_encoded)[:, 1]
            
            # محاسبه معیارهای ارزیابی برای Train
            print(f"""
                 -------------محاسبه معیارهای ارزیابی برای Train : {model_name} + {method_name}---------------------- 
                'AUC-ROC': {roc_auc_score(y_train_encoded, y_train_proba)},
                <---classification_report--->
                {classification_report(y_train_encoded, y_train_pred)}
                

     
                 -----------محاسبه معیارهای ارزیابی برای Validation : {model_name} + {method_name}-------------
                'AUC-ROC': {roc_auc_score(y_val_encoded, y_val_proba)},
                <---classification_report--->
                {classification_report(y_val_encoded, y_val_pred)}
                
        
 
                --------------محاسبه معیارهای ارزیابی برای Test : {model_name} + {method_name}--------------------- 
                'AUC-ROC': {roc_auc_score(y_test_encoded, y_test_proba)},

                <---classification_report--->
                {classification_report(y_test_encoded, y_test_pred)}
__________________________________________________________________________________________________________________________
            """)

    return results
````

````python
models = {
    'XGBoost': XGBClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'LightGBM': LGBMClassifier(random_state=42),
    "RandomForestClassifier":RandomForestClassifier(random_state=42)
}

sampling_methods = {
    'SMOTE': SMOTE(random_state=42),
    'SMOTETomek': SMOTETomek(random_state=42),
    'SMOTEENN': SMOTEENN(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'RandomUnderSampler': RandomUnderSampler(random_state=42)
}
````

````python
# TODO :  تعریف مدل‌هاو آموزش مدل ها

# فراخوانی تابع
results = data_balancing(
    methods=sampling_methods,
    models=models,
    X_train_encoded=X_train_encoded,
    X_val_encoded=X_val_encoded,
    X_test_encoded=X_test_encoded,
    y_train_encoded=y_train,
    y_val_encoded=y_val,
    y_test_encoded=y_test
)
````

````python
from imblearn.pipeline import Pipeline  
from sklearn.model_selection import RandomizedSearchCV
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import GradientBoostingClassifier


RANDOM_STATE = 42

# 1. ایجاد Pipeline با imblearn
pipeline_best_model = Pipeline([
    ('transformer', transformer),  # ترجیح بر این شد که در پاپلاین نوشته شود که بعدا در پیکل ذخیره شود
    ('sampler',  RandomUnderSampler(random_state=RANDOM_STATE)),
    ('model', GradientBoostingClassifier(random_state=RANDOM_STATE))
])

# 2. تعریف فضای پارامترها
param_dist_gbc = {
    'model__n_estimators': [20, 50, 100],
    'model__learning_rate': [0.01, 0.05, 0.07],
    'model__max_depth': [2, 4],
    'model__min_samples_split': [3, 5, 7],
    'model__subsample': [0.5, 0.7, 0.9],
    'model__max_features': ['sqrt', 'log2', None]
}

# 3. تنظیم RandomizedSearchCV
random_search_gbc = RandomizedSearchCV(
    estimator=pipeline_best_model,
    param_distributions=param_dist_gbc,
    n_iter=50,
    scoring='f1_weighted',  # تغییر معیار ارزیابی
    cv=5,
    verbose=3,
    n_jobs=-1,
    random_state=RANDOM_STATE
)

# 4. آموزش مدل
random_search_gbc.fit(X_train, y_train)

# 5. ارزیابی نتایج
best_model = random_search_gbc.best_estimator_
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:,1]

````
